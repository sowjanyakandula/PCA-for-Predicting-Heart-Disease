```python
import pandas as pd
df = pd.read_csv('heart.csv')#read the data file
```

```python
df.FastingBS.unique()
```

```python
#preprocessing
#removing the outliers i.e zscore > 3

import scipy.stats as stats
df1=df[stats.zscore(df.iloc[:,4])<3]#Cholesterol 
df1.shape
```

```python
df2=df1[stats.zscore(df1.iloc[:,3])<3]#restingBP
df2.shape
```

```python
df3=df2[stats.zscore(df2.iloc[:,7])<3]#MaxHR
df3.shape
```

```python
df4=df3[stats.zscore(df3.iloc[:,9])<3]#Oldpeak
df4.shape
```

```python
df5=df4[stats.zscore(df4.iloc[:,5])<3]#fastingBS
df5.shape
```

```python
df6=df5[stats.zscore(df5.iloc[:,0])<3]#Age
df6.shape
```

```python
y=df6.HeartDisease
x=df6.iloc[:,0:11]
x
```

```python
#Encoding Categorical data

#labelencoding
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
x['RestingECG'] = labelencoder.fit_transform(x['RestingECG'])
x['ST_Slope'] = labelencoder.fit_transform(x['ST_Slope'])
x['ExerciseAngina'] = labelencoder.fit_transform(x['ExerciseAngina'])
x
```

```python
#one hot encoding
import numpy as np
x=pd.get_dummies(x, columns=["Sex","ChestPainType"])
```

```python
x
```

```python
#scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
#Clasification before applying PCA

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2,random_state = 30)

model =  LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
#PCA

from sklearn.decomposition import PCA

pca = PCA(0.95)#retain 95% of info,variation,upto the system to decide on No of components(calculates new columns)
x_pca = pca.fit_transform(x_scaled)
x_pca.shape
```

```python
x_pca
pca.explained_variance_ratio_
pca.n_components_
```

```python
#Clasification after applying PCA

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.2,random_state = 30)
model =  LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
#PCA with number of components specified explicity

pca = PCA(n_components=3)#retain 95% of info,variation,upto the system to decide on No of components(calculates new columns)
x_pca = pca.fit_transform(x_scaled)
x_pca.shape
```

```python
#Clasification after applying PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.2,random_state = 30)
model =  LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
